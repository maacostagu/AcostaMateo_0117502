```python
#%% 3                                  
def multiplo(valor, multiplo):
    """
    Funcion para calcular si el numero es multiplo
    utilizando el modulo de la division
    """
    residuo = valor % multiplo   
    if residuo == 0:      #si el resto entre los dos anteriores es 0 significa
        #que uno es multiplo del otro.
        return True
    else:
        return False
 
# lista que contendra los valores multiples.
multiplos_3=[]  
multiplos_4=[]
 
# bucle del 1 al 100.
for i in range(1,101):
 
    if multiplo(i,3):    
        multiplos_3.append(i)  # mediante append agregamos cada numero que 
        #este en el rango solicitado y cumpla la condicion a la lista de 
        #multiplos de 3. 
 
    if multiplo(i,4):
        multiplos_4.append(i)
#se imprimen los resultados.
print ("Los multiplos de 3 son:", multiplos_3)   
print( "")
print( "Los multiplos de 4 son:", multiplos_4)
```

```python
#%% 5
import keyword 
def leer_frase():
    """Se define una función que haga que el texto digitado por el usuario se 
    almacene en una variable global"""
    global txt  # se almacena la variable txt como global, esto para que pueda
    #ser utilizada en las demás funciones
    txt = (input('ingrese un texto:'))  #Alacenamos el texto ingresado como 
    #variable global   
    
def contar_letras():
    """Se define un función que me cuenta las tildes, vocales, mayusculas, 
    digitos, espacios y palabras reservadas de una cadena de texto"""
    #se crean todos los contadores requeridos igualados a 0.
    conta = 0      
    conta_tildes=0
    conta_vocal=0
    conta_mayus=0
    conta_digit=0
    conta_space=0
    conta_reservadas=0
    for i in txt:
        if (i.isalpha()): #con is alpha nos dice si lo que esta en i es 
            #alfabetico
            conta += 1
        if i=='a' or i=='e' or i=='i' or i=='o' or i=='u': #aqui le decimos al 
            #programa que cada que encuentre una de estas letras(vocales) 
            #aunmente el contador
            conta_vocal +=1
        if (i.isupper()): #con isuper sabemos si hay mayusculas
            conta_mayus +=1
            
    for j in txt:
        if j=='á'or j=='é' or j=='í' or j=='ú' or j=='ó':  #aqui sabremos 
            #cuantas tildes hay y las coloco solo en las vocales por que 
            #siempre las tildes van en las vocales
            conta_tildes +=1
    
    for x in txt:
        if (x.isdigit()): #con is digit sabremos si lo ingresado contiene 
            #digitos (números) 
            conta_digit +=1
    
    for y in txt:
        if (y.isspace()): #aqui sabremos cuantos espacios hay 
            conta_space +=1
            
    for z in txt:
        if keyword.iskeyword(z): #este comando nos permite conocer cuantas 
            #palabras reservadas hay en el texto
            conta_reservadas +=1
    
            
    print('La cantidad de letras es:',conta)
    print('la cantidad de vocales es:',conta_vocal)
    print('la cantidad de mayusculas es:',conta_mayus)
    print('la cantidad de tildes es:',conta_tildes)
    print('la cantidad de digitos es:',conta_digit)
    print('la cantidad de espacios es:',conta_space)    
    print('la cantidad de palabras reservadas es:',conta_reservadas)
```

```python
#%% 9
def par(x):
    'funcion que indica si un numero es par'
    return x%2==0
def suc(n):
    'funcion que me retorna el siguiente numero de la sucesion'
    #condicionles si el numero es par o impar
    if par(n):
        return n/2
    else:
        return 3*n+1
def sucesion():
    while True: #se verifican os datos iniciales y que sean admisibles 
        try:
            Ui=int(input('digite el primer numero:'))
            ni=int(input('cuantos terminos quiere:'))
            break
        except ValueError:
            print('el numero no es valido,intente nuevamente')
    a=[Ui] #se crea la lista inicial con el valor de Uo
    for i in range(ni-1):  
        a.append(int(suc(a[i]))) #se evalua la sucesion hasta la cantidad de terminos deseados 
    print(f'los terminos de la sucesin son:\n{a}')
sucesion()
```

```python
#%%13
while True: #Si no se cumplen los ciclos se volverá a ejecutar el programa 
    #automaticamente.
    try:
        
        def años_bisiestos(año1,año2,x = 0):
            """Definimos una función para contar los años bisiestos de un año
            a otro"""
           
            for i in range(año1,año2+1):#Iteramos mediante un ciclo for para
                #hallar que resto entre los números debajo son iguales a 0 
                #para cumplir con la ecuación para calcular un año bisiesto
                if i%400==0 or (i%100!=0 and i%4==0):
                    x += 1 # si se cumple la condición de if añadimos una
                    #unidad al contador.
            print(f'Los años bisiestos son {x}.') #Imprimimos el total de los 
            #años bisiestos al salir del ciclo 2
                    
        año_1=int(input('ingrese año 1:')) #solicitamos al usuario ingresar el 
        #año del que quiere que el programa empiece a iterar y lo almacenamos 
        #en una variable.
        año_2=int(input('ingrese año 2:')) # Solisitamos al usuario ingresar el
        #año hasta donde quiere que el programa deje de iterar y lo almacenamos
        #en una variable.
        
        if año_1 < 0 or año_2 < 0: # si las variables son negativas:
            print('No puede ingresar años negativos')
        else: # de lo contrario ejecutar la función automáticamente
            años_bisiestos(año_1,año_2)
        break # cerramos el ciclo
            
    except ValueError: # Le informamos al usuario que ingresó un carácter 
        #erróneo
        print('No puede ingresar letras')  
```

```python
#%%14

import random

def sumar():
    "se define la funcion que sumara los numeros positivos"
    print('escriba el resultado de las siguientes operaciones:')
    #se crea el contador
    correcta=0
    #se crea un ciclo que funcionara hasta que el usuario acierte 5 sumas
    while correcta < 5:
        #se generan numeros aleatorios en un rango entre 1 y 100
        a= random.randrange(1,101) 
        b= random.randrange(1,101)
        respuesta= int(input(f'{a} + {b} = ')) 
        #se cran los condicionales para la respuesta que de el usuario 
        if respuesta==a + b: 
            correcta += 1
            print('respuesta correcta')
        else:
            print('respuesta incorrecta')
            break
        print()
    print('fin')
```
